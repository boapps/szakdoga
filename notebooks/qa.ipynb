```python
from datasets import load_dataset

dataset = load_dataset('boapps/kmdb_base')
```

```python
dataset['train']['text'][:5]
```

```python
from llama_index import (
    KeywordTableIndex,
    SimpleDirectoryReader,
    LLMPredictor,
    ServiceContext
)
from llama_index.llms import OpenAI
# alternatively
# from langchain.llms import ...

from llama_index import Document

documents = [Document(text=t) for t in dataset['train']['text'][:5]]

# define LLM
llm = OpenAI(temperature=0.1, model="text-davinci-003")
service_context = ServiceContext.from_defaults(llm=llm)

# build index
index = KeywordTableIndex.from_documents(documents, service_context=service_context)

# get response from query
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("Mi volt a szerelmi szál a Zuchlag ügyben?")
```
