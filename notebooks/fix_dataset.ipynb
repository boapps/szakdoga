```python
def process(row):
    if row['kmonitor_tags'] is None:
        row['kmonitor_tags'] = []
    else:
        row['kmonitor_tags'] = list(row['kmonitor_tags'].split(','))
    if row['keywords'] is None:
        row['keywords'] = []
    else:
        row['keywords'] = list(row['keywords'].split(','))
    if row['institutes'] is None:
        row['institutes'] = []
    else:
        row['institutes'] = list(row['institutes'].split(','))
    if row['people'] is None:
        row['people'] = []
    else:
        row['people'] = list(row['people'].split(','))
    if row['authors'] is None:
        row['authors'] = []
    else:
        row['authors'] = list(row['authors'].split(','))

    row['kmonitor_places'] = []
    row['kmonitor_collections'] = []
    return row

dataset = dataset.map(process)
```

```python
dataset = dataset.rename_column("institutes", "kmonitor_institutes")
dataset = dataset.rename_column("people", "kmonitor_people")
```

```python
def process(row):
    tags = row['kmonitor_tags']
    for tag in tags:
        if tag not in all_tags and tag not in all_collections:
            row['kmonitor_places'].append(tag)
            row['kmonitor_tags'] = [a for a in row['kmonitor_tags'] if a != tag]
        if tag not in all_tags and tag in all_collections:
            row['kmonitor_collections'].append(tag)
            row['kmonitor_tags'] = [a for a in row['kmonitor_tags'] if a != tag]
    return row

dataset = dataset.map(process)
```
